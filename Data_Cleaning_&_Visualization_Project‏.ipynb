```python
#الاستيراد
import seaborn as sns
import pandas as pd
```

```python
# قراءة ملف البيانات
data = pd.read_excel('train.xls')
data
```

```python
# عرض أول 5 صفوف في البيانات
print(data.head())

# معلومات عن البيانات
print(data.info())

# إحصاءات وملخصات عن البيانات العددية
print(data.describe())
```

```python
#:معالجة القيم المفقودة

# عدد القيم المفقودة في كل عمود
print(data.isnull().sum())

# ملء القيم المفقودة في عمود العمر بالمتوسط
data['Age'].fillna(data['Age'].mean(), inplace=True)

# ملء القيم المفقودة في عمود المحطة بالقيمة الأكثر تكراراً
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
# عرض أسماء الأعمدة المتوفرة في المجموعة البيانات
print(data.columns)

# تحديث أسماء الأعمدة في الأمر data.drop
data.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
```
